```python
import matplotlib.pyplot as plt

# Sorting the items to display them in a visually appealing manner
sorted_circles = sorted(circles.items(), key=lambda x: -x[1])

# Setup the plot
fig, ax = plt.subplots(figsize=(10, 2))  # Wider figure to accommodate all circles
ax.axis('off')  # Hide axes

# Colors and additional settings
colors = plt.get_cmap('Blues')
base_size = 1000  # Base size for circles
max_val = max(circles.values())

# Plot each circle
for i, (name, val) in enumerate(sorted_circles):
    ax.scatter([i], [0], s=(val/max_val) * base_size, color=colors(val / max_val), alpha=0.6, edgecolors='none')
    ax.text(i, 0, name, ha='center', va='center', fontweight='bold' if name == 'ECloudGen' else 'normal')

plt.show()
```
